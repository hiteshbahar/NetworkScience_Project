```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
%matplotlib inline
```

```python
Patches = 40   # Number of patches (hospitals)
P_ext = 0.005    # Probability of extinction (e)
P_col = 0.01   # Probability of transfer (c)
P_init = 0.8   # Probability that a patch will be occupied at the beginning
Distance = 3  # An arbitrary parameter to determine which patches are connected
Edges = 0.25
```

```python
class patch:
    def __init__(self,status=0):
        self.status = status
        #self.pos = pos
    def __str__(self):
        return(str(self.status))
```

```python
# Start the simulation
for n in G.nodes():
    if (Stat == 1 and np.random.uniform() < P_ext):
        Stat = 0
```

```python
for n in G.nodes():
    if Stat == 1:
        neighb = G[n] # That's it, a list of the neighbors
        for nei in neighb:
            if Stat == 0:
                if np.random.uniform() < P_col:
                    Stat = 1
                    break
```

```python
for timestep in range(2000):
    # Check for extinctions
    for n in G.nodes():
        if (Stat == 1 and np.random.uniform() < P_ext):
            Stat = 0
    # Check for invasions
    for n in G.nodes():
        if Stat == 1:
            neighb = G[n] # That's it, a list of the neighbors
            for nei in neighb:
                print(nei)
                #if Stat == 0:
                    #if np.random.uniform() < P_col:
                        #Stat = 1
                        #break
```
